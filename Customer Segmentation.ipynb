# Customer Life Cycle RFM ANALYSIS

```python
# Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
#read_dataset
retail = pd.read_csv("retail_clean.csv")
retail.head(3)
```

```python
retail.rename(columns={"InvoiceDate":"date"},inplace=True)
```

```python
retail['date']  = pd.to_datetime(retail['date'])
```

```python
retail.info()
```

```python
retail['date']= retail['date'].dt.strftime('%Y-%m-%d')
```

```python
retail['date']= pd.to_datetime(retail['date'])
```

```python
retail.info()
```

```python
max_date= retail['date'].max()
max_date
```

### Recency

```python
retail.columns
```

```python
customers_recency= retail.groupby('Customer ID').agg(last_date= ('date','max')).reset_index()
```

```python
customers_recency
```

```python
customers_recency['recency']= max_date- customers_recency['last_date']
```

```python
customers_recency
```

```python
customers_recency['recency']=customers_recency['recency'].astype('string').str.replace('days','').astype(int)
```

```python
customers_recency
```


#### frequency

```python
freq2= retail.groupby('Customer ID').date.count().reset_index()
freq2
```

```python
freq2.columns= ['Customer ID','frequency']
freq2
```


### Monetary value

```python
retail.columns
```

```python

monet1= retail.groupby(['Customer ID','Invoice']).agg(revenue= ('Revenue','sum')).reset_index()
monet1
```

```python
monet2= monet1.groupby('Customer ID').agg(monetary= ('revenue','mean')).reset_index()
monet2
```

### Ranking

```python
customers_recency['rank_recency']= customers_recency['recency'].rank(pct=True)
```

```python
freq2['freq_ranking']=freq2['frequency'].rank(ascending=False,pct=True)
```

```python
monet2['rank_monet']= monet2['monetary'].rank(ascending=False,pct=True)
```

```python
customers_recency
```

```python
freq2
```

```python
monet2
```

```python
all_data= pd.merge(customers_recency, freq2,on='Customer ID')
```

```python
all_data
```

```python
monet2
```

```python
all_data= pd.merge(all_data,monet2,on='Customer ID')
```

```python
all_data = all_data.loc[:,~all_data.columns.duplicated()]
all_data.rename(columns={
    'monetary_x':'monetary',
    'rank_monet_x':'rank_monet'
},inplace=True)

all_data = all_data[['Customer ID','recency','frequency','monetary','rank_recency','freq_ranking','rank_monet']]
all_data
```

# # normalize each rank with Max rank

```python
all_data['R_rank_normalized']=(all_data['rank_recency']/all_data['rank_recency'].max())*100
all_data['F_rank_normalized']=(all_data['freq_ranking']/all_data['freq_ranking'].max())*100
all_data['M_rank_normalized']=(all_data['rank_monet']/all_data['rank_monet'].max())*100


all_data
```

```python
#*Consolidated Score = 0.15*R_rank_norm + 0.28*F_rank_norm + 0.57M_rank_norm**


all_data['CLV']=0.15*all_data['R_rank_normalized']+0.28*all_data['F_rank_normalized']+0.57*all_data['M_rank_normalized']
all_data.head()
```

```python
# ## Customer segments with RFM Model
# # Segment customers based on RFM score
# 0 - 50 - Low valued customer
# 50 - 75 - Medium valued customer
# 76 - 100 - High valued customer High Value Customer: They are your loyal customers.So provide constant support via customer service
#Medium Value Customer: Send them personalized emails with offers and encourage them to shop more
#Low Value Customer: These customers are about to churn or went into a dormant stage,apply reactivation strategies for them


all_data["Customer_segment"]=np.where(all_data['CLV'] > 75 ,"High Value Customer",(np.where(all_data['CLV'] < 50 , "Low value Customer" ,"Medium Value Customer")))
all_data.head()
```
